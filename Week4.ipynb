```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=np.array([[2,22],[4,41],[6,54],[5,50],[1,12]])
plt.scatter(data[:,0],data[:,1])
```

```python
x=np.array([[1,2],[1,4],[1,6],[1,5],[1,1]])
```

```python
xTx=np.dot(x.T,x)
xTx_inv=np.linalg.inv(xTx)
w=np.dot(xTx_inv,xTx)
print(w)
```

```python
x=np.linspace(-1,10,100)
y=w[0,0]+w[0,1]*x
plt.scatter(data[:,0],data[:,1])
plt.plot(x,y)
```
